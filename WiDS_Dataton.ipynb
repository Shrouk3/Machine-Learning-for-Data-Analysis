```python
# Check for null values
null_counts = df.isnull().sum()
print("Null values in each column:")
print(null_counts[null_counts > 0])
```

```python
plt.figure(figsize=(14, 8))
null_counts[null_counts > 0].sort_values().plot(kind='barh', color='orange')
plt.xlabel("Number of Missing Values")
plt.title("Null Values Count per Column")
plt.show()
```

```python
# to draw scatter use float values
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="ELEVATION", y="january_avg_temp", hue="building_class", palette="Set2")
plt.title("January Average Temperature vs. Elevation by Building Class")
plt.xlabel("Elevation")
plt.ylabel("January Average Temperature (Â°F)")
plt.legend(title="Building Class")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y="facility_type", order=df['facility_type'].value_counts().index, palette="viridis")
plt.title("Distribution of Facility Types")
plt.xlabel("Count")
plt.ylabel("Facility Type")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["energy_star_rating"], bins=20,kde = True, color="skyblue")
plt.title("Distribution of Energy Star Ratings")
plt.xlabel("Energy Star Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="days_with_fog", y="site_eui", palette="Blues")
plt.title("Site EUI by Days with Fog")
plt.xlabel("Days with Fog")
plt.ylabel("Site EUI")
plt.show()
```
